```python
import random
round = 0
total_distance = 0

while round <3:
    round+=1
    print("This is round",round)
    lucky_number = random.randint(1,5)
    chance = 0
    while chance<3:
        IR1 = got.read_distance_data(21)
        chance+=1
        guess = int(input("Please select a number 1-5:"))
        if guess==lucky_number:
            print("You are right!")
            if IR1>30:
                distance = random.randint(5,15)
                got.mecanum_move_speed_times(0,25,distance,1)
                print("The vehicle just moved",distance,"cm.")
                total_distance+=distance
                break
            else:
                print("Good job, you've won the game!!")
                break
        else:
            print("You are wrong,try again!")
    else:
       print("This round is over,the lucky number is",lucky_number,"try again?")           
else:
    print("Game over, see you!")
    print("You vehicle moved",total_distance,"cm")
```
