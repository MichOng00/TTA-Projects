```python
def AP_reco_app():
    import time
    got.load_models(["apriltag_qrcode"])
    AP_info = got.get_apriltag_total_info()
    
    # Step 1 AP recognition
    while True:
        AP_info = got.get_apriltag_total_info()
        if not AP_info:
            got.mecanum_move_xyz(0,0,20)
        else:
            got.mecanum_stop()
            break
    print("Apriltag recognized.")
    
    # Step 2 AP centralization
    while True:
        AP_info = got.get_apriltag_total_info()
        x_coordinate = AP_info[0][1]
        if x_coordinate<330 :
            got.mecanum_move_xyz(0,0,10)
        else:
            print(x_coordinate)
            got.mecanum_stop()
            break
    print("Apriltag cetralized.")
    
    # Step 3 approach
    while True:
        AP_info = got.get_apriltag_total_info()
        if AP_info:
            x_coordinate = AP_info[0][1]
            distance = AP_info[0][6]
        if x_coordinate<290:
            got.mecanum_move_xyz(-5,5,0)
        elif x_coordinate>350: 
            got.mecanum_move_xyz(5,5,0)  
        elif distance>0.2:
            got.mecanum_move_xyz(0,10,0)
        else:
            got.mecanum_stop()
            break
    print("It is too close,stop!")
```

```python
def line_following():
    import time
    got.load_models(["line_recognition"])
    got.set_track_recognition_line(0)
    
    line_info = got.get_single_track_total_info()
    line_type = line_info[1]
    offset = line_info[0]
    
    while line_type >0:
        line_info = got.get_single_track_total_info()
        line_type = line_info[1]
        offset = line_info[0]
        degrees = int(0.25*offset)
        got.mecanum_move_xyz(0,25,degrees)
        time.sleep(0.1)
    got.mecanum_stop()
    print("I can not see the line.")

def AP_recognition():
    import time
    got.load_models(["apriltag_qrcode"])
    AP_info = got.get_apriltag_total_info()
    
    # Step 1 AP recognition
    while True:
        AP_info = got.get_apriltag_total_info()
        if not AP_info:
            got.mecanum_move_xyz(0,0,20)
        else:
            got.mecanum_stop()
            break
    print("Apriltag recognized.")
    
    # Step 2 AP centralization
    while True:
        AP_info = got.get_apriltag_total_info()
        x_coordinate = AP_info[0][1]
        if x_coordinate<330 :
            got.mecanum_move_xyz(0,0,10)
        else:
            print(x_coordinate)
            got.mecanum_stop()
            break
    print("Apriltag cetralized.")

def AP_approaching():
    # Step 3 approach
    while True:
        AP_info = got.get_apriltag_total_info()
        if AP_info:
            x_coordinate = AP_info[0][1]
            distance = AP_info[0][6]
            if x_coordinate<290:
                got.mecanum_move_xyz(-5,5,0)
            elif x_coordinate>350: 
                got.mecanum_move_xyz(5,5,0)  
            elif distance>0.2:
                got.mecanum_move_xyz(0,10,0)
            else:
                got.mecanum_stop()
                break
    print("It is too close,stop!")

def Pose_angle_right():
    import time
    got.load_models(["apriltag_qrcode"])
    AP_info = got.get_apriltag_total_info()
    pose_z = AP_info[0][9]
    
    while True:
        AP_info = got.get_apriltag_total_info()
        pose_z = AP_info[0][9]
        if pose_z>0:
            got.mecanum_move_xyz(10,0,15)
        else:
            got.mecanum_stop()
            break
    print("Apriltag is centralized squarely.")

def Pose_angle_left():
    import time
    got.load_models(["apriltag_qrcode"])
    AP_info = got.get_apriltag_total_info()
    pose_z = AP_info[0][9]
    
    while True:
        AP_info = got.get_apriltag_total_info()
        pose_z = AP_info[0][9]
        if pose_z<0:
            got.mecanum_move_xyz(-10,0,-15)
        else:
            got.mecanum_stop()
            break
    print("Apriltag is centralized squarely.")

def pick_up():
    import time
    got.mechanical_arms_restory()
    time.sleep(2)
    got.mechanical_joint_control(0,0,0,1000)
    got.mechanical_clamp_release()
    time.sleep(1.5)
    got.mechanical_joint_control(0,-5,-60,1000)
    time.sleep(2)
    got.mechanical_clamp_close()
    time.sleep(1.5)
    got.mechanical_joint_control(30,0,30,1000)
    
def put_down():
    import time
    got.mechanical_joint_control(0,-10,-50,1000)
    time.sleep(2)
    got.mechanical_clamp_release()
    time.sleep(1.5)
    got.mechanical_arms_restory()
    time.sleep(2)
    got.mechanical_clamp_close()
    time.sleep(1.5)
```
