```python
from ugot import ugot
import time

got = ugot.UGOT()

got.initialize('192.168.1.88')
```

```python
got.mecanum_stop() 
```

```python
# move forward, move backward, stop, turn left, turn right
got.load_models(["gesture"])
options = ["Cloth", "Stone", "Scissors", "OK", "Likes"]

while True:
    gesture = got.get_gesture_result()
    if gesture in options:
        # print(gesture)
        if gesture == "Stone":
            got.mecanum_move_speed(0,15)
        elif gesture == "Scissors":
            got.mecanum_move_speed(1,15)
        elif gesture == "Cloth":
            got.mecanum_stop() 
        elif gesture == "OK":
            got.mecanum_turn_speed(3,30)
        elif gesture == "Likes":
            got.mecanum_turn_speed(2,30)
```

```python
# add commands to list - now move fixed distance (cloth to activate)
got.load_models(["gesture"])
options = ["Cloth", "Stone", "Scissors", "OK", "Likes"]
threshold = 0
while True:
    go = False
    commands = []

    while not go:
        gesture = got.get_gesture_result()
        if gesture in options[1:]: # any option except cloth
            threshold = 0
            commands.append(gesture)
            got.play_audio_tts(f"Added {gesture}", voice_type=0, wait=True)
            print(commands)
        elif gesture == "Cloth":
            threshold += 1
            if threshold == 50:
                got.play_audio_tts(f"Commands activated", voice_type=0, wait=True)
                go = True

    for gesture in commands:
        if gesture == "Stone":
            got.mecanum_move_speed_times(0,30,50,1)
        elif gesture == "Scissors":
            got.mecanum_move_speed_times(1,30,50,1)
        elif gesture == "OK":
            got.mecanum_turn_speed_times(3,30,90,2)
        elif gesture == "Likes":
            got.mecanum_turn_speed_times(2,30,90,2)
```
