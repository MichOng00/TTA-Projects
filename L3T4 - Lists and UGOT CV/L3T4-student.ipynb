```python
# write a function to pick up the ball from the table / ground
def pickup():
    got.mechanical_arms_restory()
    got.mechanical_clamp_release()
    time.sleep(1)
    got.mechanical_joint_control(angle1=0, angle2=-30, angle3=-50, duration=500) # change the numbers
    time.sleep(2)
    got.mechanical_clamp_close()
    time.sleep(2)
    got.mechanical_joint_control(0, 0, 0, 500)

pickup() # don't forget to call the function

# write a function to throw the ball
def throw():
    got.mechanical_joint_control(0, 80, 80, 500)
    time.sleep(1)
    got.mechanical_joint_control(0, 0, 0, 50)
    got.mechanical_clamp_release()

throw()
```

```python
import time
from ugot import ugot
got = ugot.UGOT()
got.initialize("192.168.1.29") # your IP address!
 
got.load_models(["line_recognition", "apriltag_qrcode"])
 
def line_follow():
    got.set_track_recognition_line(line_type = 0) # single track line
 
    line_info = got.get_single_track_total_info()  # list: [offset, type, x, y]
    offset = line_info[0] # tell you how far the line is from straight
    line_type = line_info[1] # what kind of line
    
    while line_type > 0: # there is some kind of line in front of the robot
        # update information continuously - otherwise it will not change
        line_info = got.get_single_track_total_info()
        offset = line_info[0]
        line_type = line_info[1]
        
        degrees = int(offset / 4) # prevent it from turning too much
        # convert from float to int - otherwise TypeError
        got.mecanum_move_xyz(0, 20, degrees)
        time.sleep(0.1)
    
    got.mecanum_stop()
 
# line_follow()
 
def ap_recog():    
    # step 1: move right until apriltag recognised
    while True:
        AP_info = got.get_apriltag_total_info()
        if not AP_info: # shorter than 'if AP_info == []:'
            got.mecanum_move_xyz(10, 0, 0) # move right
        else:
            got.mecanum_stop()
            break 
    print("Apriltag recognized.")
 
    # step 2: centralize - keep moving right until tag is in the centre
    while True:
        AP_info = got.get_apriltag_total_info()
        if AP_info:
            x_coord = AP_info[0][1] # x-coordinate of the centre of apriltag
            if x_coord < 330:
                got.mecanum_move_xyz(10, 0, 0)
            else:
                got.mecanum_stop()
                break 
    print("Apriltag centralised.")
 
    # step 3: approach apriltag to pick it up
    while True:
        AP_info = got.get_apriltag_total_info()
        if AP_info:
            x_coord = AP_info[0][1]
            dist = AP_info[0][6]
            if x_coord < 290:
                 got.mecanum_move_xyz(-3, 3, 0) # move left and forward
            elif x_coord > 350:
                got.mecanum_move_xyz(3, 3, 0) # move right and forward
            elif dist > 0.12: 
                got.mecanum_move_xyz(0, 3, 0) # move straight forward
            else:
                got.mecanum_stop()
                break
        else:
            got.mecanum_stop()
    print("Stopped.")
 
# ap_recog()
 
def pickup_ap():
    got.mechanical_clamp_release()
    time.sleep(1)
    got.mechanical_joint_control(0, 0, -70, 800) #down - for apriltag
    time.sleep(1)
    got.mechanical_clamp_close()
    time.sleep(2)
    got.mechanical_joint_control(0, 30, -50, 800) #up
 
def put_ap():
    got.mechanical_joint_control(-90, 30, -50, 800)
    time.sleep(1)
    got.mechanical_joint_control(-90, -20, -30, 800)
    time.sleep(1)
    got.mechanical_clamp_release()
    time.sleep(2)
```

The robot should scan for the apriltag until it recognises it, then approach and pick it up.
The apriltag will be on the line.
Once the robot picks up the apriltag, it should start following the line.
The robot should put down the apriltag at the end of the line.

```python
# MAIN CODE
# USE YOUR FUNCTIONS and time.sleep(1)
# line_follow(), ap_recog(), pickup_ap(), put_ap()
got.mechanical_joint_control(0, 0, 0, 500)
ap_recog()
time.sleep(1)
pickup_ap()
time.sleep(1)
line_follow()
time.sleep(1)
put_ap()
```
