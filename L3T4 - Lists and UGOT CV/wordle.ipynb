```python
from ugot import ugot
got = ugot.UGOT()
got.initialize('192.168.1.88')
```

```python
import time
secret = "robot" # make sure lowercase
got.load_models(["word_recognition"])
guessed = False

while not guessed:
    result = ""
    player_guess = got.get_words_result().lower().strip()
    #player_guess = input()
    if len(player_guess) == len(secret):
        print(player_guess)
        for i in range(len(secret)):
            if secret[i] == player_guess[i]:
                result += "ðŸŸ©"
                got.show_light_rgb_effect(0, 255, 0, 0)
                time.sleep(1)
                got.turn_off_lights()
                time.sleep(1)
            elif player_guess[i] in secret:
                result += "ðŸŸ¨"
                got.show_light_rgb_effect(255, 120, 0, 0)
                time.sleep(1)
                got.turn_off_lights()
                time.sleep(1)
            else:
                result  += "â¬›"
                got.show_light_rgb_effect(255, 0, 0, 0)
                time.sleep(1)
                got.turn_off_lights()
                time.sleep(1)
        print(result)
        if player_guess == secret:
            guessed = True
        time.sleep(1)
```
