```python
import pygame

# Initialize Pygame
pygame.init()

# Set up the screen
screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Hello, Pygame!")

# Colors
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# Game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    # Fill screen with white
    screen.fill(WHITE)
    
    # Draw shapes
    pygame.draw.rect(screen, RED, (100, 100, 200, 100)) # surface, color, position (left, top, width, height)
    pygame.draw.circle(screen, (0, 0, 255), (400, 300), 50) # surface, color, center, radius

    # Update display
    pygame.display.flip()

quit()
```
