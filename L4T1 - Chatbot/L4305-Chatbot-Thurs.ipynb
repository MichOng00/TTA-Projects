```python
# polygon
def polygon(n, length):
    # n is number of sides
    angle = 360 / n
    for i in range(n):
        forward(length)
        right(angle)

make_turtle(width=300,height=300,delay=0.1)
polygon(50, 10)
```

```python
import math

def circle(radius):
    circumference = 2 * math.pi * radius
    n = 40 # how precise the circle is
    length = circumference / n
    polygon(n,length)

make_turtle(delay = 0.01, height = 200)
circle(40)
```

```python
def multi_circle(n): #n is number of circles
    angle = 360/n
    for i in range(n):
        set_color(colors[i%len(colors)])
        circle(30)
        left(angle)

make_turtle(delay = 0.001)
multi_circle(24)
```

```python
import random

colors = 'red OrangeRed Tomato DarkOrange Coral Orange'.lower().split()

def random_painting(steps, length):
    for i in range(steps):
        set_color(random.choice(colors))
        forward(length)
        turn_angle = random.choice([0, 90, 180, 270])
        left(turn_angle)

make_turtle(delay=0.02, width=600, height=600)
random_painting(2000, 10)
```

```python
# pastebin.com/wi8vFdvw
from textblob import TextBlob
from textblob import Word

def mood_chatbot():
    print("Hi! I'm your friendly chatbot :)")
    print("How are you feeling today?")
    
    # Get user input
    user_response = input("You: ")
    
    # Analyze sentiment using TextBlob
    blob = TextBlob(user_response)
    sentiment = blob.sentiment.polarity
    print(f'Sentiment score: {sentiment}')
    
    # Evaluate the sentiment and respond accordingly
    if sentiment > 0.5:
        print("Chatbot: That's great to hear! You seem really happy today!")
    elif sentiment > 0:
        print("Chatbot: Sounds like you're doing pretty well!")
    elif sentiment == 0:
        print("Chatbot: Hmm, I can't quite tell how you're feeling. Neutral, maybe?")
    else:
        print("Chatbot: Oh no! It seems like you're feeling a bit down. I'm here if you want to talk.")

    print("Chatbot: Thanks for sharing! Have a great day!")

# Run the chatbot
# mood_chatbot()
```

```python
import random

def recommend_chatbot():
    activities = {"positive": ["download games", "sing"],
                  "neutral": ["watch youtube at youtube.com", "play games"],
                  "negative": ["sleep", "play with lightning"]
                 }
    print("Hi! I'll recommend something to do based on your mood.")
    print("How are you feeling today?")
    
    # Get user input
    user_response = input("You: ")
    
    # Analyze sentiment using TextBlob
    blob = TextBlob(user_response)
    sentiment = blob.sentiment.polarity
    print(f'Sentiment score: {sentiment}')
    
    # Evaluate the sentiment and respond accordingly
    if sentiment > 0.5:
        activity = random.choice(activities["positive"])
        print(f"Chatbot: That's great to hear! You seem really happy today! Why not {activity}?")
    elif sentiment >= 0:
        activity = random.choice(activities["neutral"])
        print(f"Chatbot: Hmm, I can't quite tell how you're feeling. Neutral, maybe? Try to {activity}.")
    else:
        activity = random.choice(activities["negative"])
        print(f"Chatbot: Oh no! It seems like you're feeling a bit down. Maybe it would help to {activity}.")
        
    print("Chatbot: Thanks for sharing! Have a great day!")

# recommend_chatbot()
```

```python
def error_checking_chatbot():
    print("Let's chat! If I notice any spelling mistakes, I'll suggest corrections.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Chatbot: Goodbye!")
            break
            
        blob = TextBlob(user_input)
        corrected_input = blob.correct()
        
        # Check if correction is needed
        if str(blob) != str(corrected_input):
            print(f"Chatbot: Did you mean '{corrected_input}'?")
            confirmation = input("Type 'yes' if correct, or 'no' to keep your original text: ").strip().lower()
            if confirmation == "yes":
                print("Chatbot: Great! Let's continue.")
            else:
                print("Chatbot: No problem, I'll keep your original input.")
        else:
            print("Chatbot: No errors detected.")

        print("Chatbot: Thanks for sharing! What else is on your mind?")

# error_checking_chatbot()        
```

```python
# pastebin.com/uZzAS089
from ugot import ugot
got = ugot.UGOT()
got.initialize('192.168.1.197')

def ugot_moves():
    commanding = True
    
    command_list = ['forward', 'backward', 'left', 'right']
    
    print("Welcome to the UGOT Command Center! Type 'exit' to quit. Type 'help' for a list of commands.")
    
    while commanding:
        print("What are your orders?")
        command = input("> ").lower() # forward 10 left 90
		### convert to list
        command = command.split(" ")
        
        for i, value in enumerate(command):

            if value == 'exit':
                ### change flag
                commanding = False
                print('Exiting robot...')
    
            if value == 'help':
                print('Here are the commands: ')
                for j in command_list:
                    print(f'- {j}')
    
            if value == 'forward':
                print('Moving...')
                got.transform_move_speed_times(0, 30, int(command[i+1]), 1)
            
            if value == 'backward':
                print('Moving...')
                got.transform_move_speed_times(1, 30, int(command[i+1]), 1)
    
            if value == 'left':
                print('Moving...')
                got.transform_turn_speed_times(2, 30, int(command[i+1]), 2)
    
            if value == 'right':
                print('Moving...')
                got.transform_turn_speed_times(3, 30, int(command[i+1]), 2)

### call the function to test!
# ugot_moves()
```

```python
import random

def guess_the_number():
    # generate random number
    secret_num = random.randint(1,1000) # TODO: change back to 100
    attempts = 0
    print("I've picked a number between 1 and 100. Can you guess it?")
    while True:
        try:
            guess = int(input("Your guess: "))
            attempts += 1 # this means attempts = attempts + 1
            if guess == secret_num:
                print(f"Congratulations! You guessed it in {attempts} attempts.")
                break
            # provide hints
            elif guess < secret_num:
                print("Too low!")
            else:
                print("Too high!")
        except ValueError:
            print("Please enter a valid number.")

# guess_the_number()
```

```python
test_users = {"feng qi": "not telling you", "zi qian" : "not telling you times 2"}

# add user
def add_user(users):
    adding = True
    while adding:
        new_username = input("Add username here: ")
        if new_username in users:
            print("Error: user already exists!")
        else:
            new_password = input("Add password here :")
            users[new_username] = new_password
            adding = False

add_user(test_users)
```

```python
def calculator():
    print("Welcome to the Python calculator! You can perform the following operations:")
    print("Addition (+), subtraction (-), division (/), multiplication (*), power (^)")
    print("Type \'exit\' to quit.")

    operations = {
        "+" : lambda x, y : x + y,
        "-" : lambda x, y : x - y,
        "/" : lambda x, y : x / y if y != 0 else "Error: division by zero",
        "*" : lambda x, y : x * y,
        # extra: power
        "^" : lambda x, y : x**y #caret
    }
    while True:
        operation = input("Enter the operation (+, -, /, *, ^) or type \'exit\': ").strip()
        if operation.lower() == "exit":
            print("Goodbye!")
            break
        if operation not in operations:
            print("Invalid operation.")
            continue # go back to start of while loop 
        try:
            num1 = float(input("First number: "))
            num2 = float(input("Second number: "))
            result = operations[operation](num1, num2)
            print(f"The result is {result}.")
        except ValueError:
            print("Invalid input. Please enter numeric values.")

# calculator()
```
