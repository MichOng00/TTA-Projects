```python
from ugot import ugot
got = ugot.UGOT()
got.initialize('192.168.1.32')
```

```python
def ugot_moves():
    commanding = True
    command_list = ['forward', 'backward', 'left', 'right']
    print('Welcome to the UGOT command centre! Type "exit" to quit. Type "help" for a list of commands.')
    
    while commanding:
        print('What are your orders?')
        command = input('> ').lower()
        command = command.split(' ')

        for i, value in enumerate(command):
            if value == 'exit':
                commanding = False
                print('Exiting robot...')
                
            if value == 'help':
                print('Here are the available commands: ')
                for j in command_list:
                    print(f'- {j}')

            if value == 'forward':
                print('Moving...')
                got.mecanum_move_speed_times(0, 30, int(command[i+1]), 1)

            if value == 'backward':
                print('Moving...')
                got.mecanum_move_speed_times(1, 30, int(command[i+1]), 1)

            if value == 'left':
                print('Moving...')
                got.mecanum_turn_speed_times(2, 30, int(command[i+1]), 2)
                
            if value == 'right':
                print('Moving...')
                got.mecanum_turn_speed_times(3, 30, int(command[i+1]), 2)

ugot_moves()
```
