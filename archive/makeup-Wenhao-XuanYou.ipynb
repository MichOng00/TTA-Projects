```python
from ugot import ugot
got = ugot.UGOT()
got.initialize('192.168.1.196')

import time
```

```python
def hold():
    got.mechanical_arms_restory()
    got.mechanical_clamp_release()
    time.sleep(3)
    got.mechanical_clamp_close()

hold()
```

```python
def throw():
    got.mechanical_joint_control(0, 100, 0, 500)
    time.sleep(2)
    got.mechanical_joint_control(0, 45, 0, 20)
    got.mechanical_clamp_release()
    # joint 1: -90 to 90
    # joint 2: -80 to 110
    # joint 3: -90, 90
    # duration: 20 to 5000 milliseconds
```
